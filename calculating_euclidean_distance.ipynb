```python
import math
import pandas as pd

def calculate_euclidean_distance(v1, v2):
    squared_diffs = sum((p - q) ** 2 for p, q in zip(v1, v2))
    return math.sqrt(squared_diffs)
```

```python
# Data rows(first 5 are each vector's value, 6th is forecast)
main_data = [
        [1.036, 0.281, -0.347, 0.877, 0.271, 0.245],
        [0.281, -0.347, 0.877, 0.271, 0.245, -0.481],
        [-0.347, 0.877, 0.271, 0.245, -0.481, 0.271],
        [0.877, 0.271, 0.245, -0.481, 0.271, 0.45],
        [0.271, 0.245, -0.481, 0.271, 0.45, 0.334],
        [0.245, -0.481, 0.271, 0.45, 0.334, 0.016],
        [-0.481, 0.271, 0.45, 0.334, 0.016, 0.65],
        [0.271, 0.45, 0.334, 0.016, 0.65, 0.168]
    ]

# Base Vector B
base_vector = [0.45, 0.98, 0.29, 0.59, 0.38]

# Threshold Value

THRESHOLD = 1.3

accepted_forecasts = []
```

```python
def main():
    print(f"{'Vectors':<10} {'Euclidean Dist':<20} {'Forecast':<10} {'Status'}")
    print("-" * 50)

    for i, row in enumerate(main_data):
        # Extract Vector (first 5 values) and Forecast (6th value)
        current_vector = row[:5]
        forecast_value = row[5]

        # Calculate Distance
        dist = calculate_euclidean_distance(current_vector, base_vector)

        # Check Threshold
        status = "Rejected"
        if dist < THRESHOLD:
            status = "ACCEPTED"
            accepted_forecasts.append(forecast_value)

        print(f"{i+1:<10} {dist:<20.5f} {forecast_value:<10} {status}")
```

Saving the main-data table in csv formate

```python
# Saving main_data in CSV
columns = ['p1', 'p2', 'p3', 'p4', 'p5', 'true_next_value']
df_data = pd.DataFrame(main_data, columns=columns)

# Save Input Data to CSV
df_data.to_csv('main_data.csv', index_label='Vectors')
print("Saved 'main_data.csv'.")
```

```python
#running main fucntion
main()
```

Calculate The Outcomes How many time positive and negative value Occurs.

```python
print("\n" + "="*30)
print("STATISTICS ON ACCEPTED FORECASTS")
print("="*30)

if not accepted_forecasts:
    print(f"No vectors had a distance lower than the threshold ({THRESHOLD}).")
else:
    positive_count = sum(1 for v in accepted_forecasts if v > 0)
    negative_count = sum(1 for v in accepted_forecasts if v < 0)
    total_accepted = len(accepted_forecasts)

    pos_pct = (positive_count / total_accepted) * 100
    neg_pct = (negative_count / total_accepted) * 100

    print(f"Total Accepted: {total_accepted}")
    print(f"Positive Outcomes: {pos_pct:.2f}% ({positive_count}/{total_accepted})")
    print(f"Negative Outcomes: {neg_pct:.2f}% ({negative_count}/{total_accepted})")
    df_results = pd.DataFrame(accepted_forecasts)
```

```python
# Save Analysis Results to CSV
df_results.to_csv('forecast_table.csv', index=False)
print("Saved 'forecast_table.csv'.")
```
