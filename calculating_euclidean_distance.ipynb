```python
import math


# Vector A: First 5 percentage difference value
vec_a = [1.036, 0.281, -0.347, 0.877, 0.271]

# Vector B: skip the first value and take next five
vec_b = [0.281, -0.347, 0.877, 0.271, 0.245]

print(f"Vector A: {vec_a}")
print(f"Vector B: {vec_b}")
print("-" * 30)

# Calculate the Differences (A - B)
differences = []
for i in range(5):
    diff = vec_a[i] - vec_b[i]
    differences.append(diff)
print(f"Differences: {differences}")

# Square the Differences (A - B)^2
squared_diffs = []
for val in differences:
    sq = val ** 2
    squared_diffs.append(sq)
print(f"Squared Diffs: {squared_diffs}")


sum_sq = sum(squared_diffs)
print(f"Sum of Squares: {sum_sq:.8f}")

euclidean_dist = math.sqrt(sum_sq)

print("-" * 30)
print(f"Final Euclidean Distance: {euclidean_dist:.2f}")
```
